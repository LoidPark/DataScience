## 6.시각화 톺아보기
- Matplotlib는 Python에서 정적, 애니메이션 및 대화형 시각화를 구현하기 위한 대표적인 라이브러리입니다. 
- 설치 : 앨리스 플랫폼에 이미 설치 되어 있습니다.
- Matplotlib의 pyplot이라는 모듈을 사용해서 그래프를 그립니다. 모듈 내 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있습니다.




### 6.1 기본적인 사용법

간단한 리스트를 넣으면 x축을 0부터 시작하는 인덱스 ([0,1,2,3,...]), y축을 리스트의 값으로 하는 선 그래프가 그려집니다.

show() 함수는 그래프를 화면에 보여지게 합니다.

```python
plt.plot([1, 2, 3, 5])
plt.show()
```

x축을 따로 설정하고 싶으면 plot() 함수 내에 첫 번째 인자로 넣어 줍니다.

```python
# 첫 번째 인자는 x축, 두 번째 인자는 y축으로
plt.plot([1, 2, 3, 4], [1, 3, 2, 4])
plt.show()
```

축 범위를 지정할 때는 plt.axis()를 사용합니다.

plt.axis([xmin, xmax, ymin, ymax])

```python
plt.plot([1, 2, 3, 4], [1, 3, 2, 4])
plt.axis([0, 6, 0, 20])
plt.show()
```

여러 개의 그래프를 그릴 때는 (x1,y1), (x2,y2)... 순서대로 적어 주면 됩니다.

```python
plt.plot([1, 2, 3, 4], [1, 3, 2, 4], [1, 2, 3, 4], [5, 6, 5, 6])
plt.show()
```

### 6.3 레이블과 범례 꾸미기

하나의 데이터를 가지고 레이블과 범례를 넣고 꾸며 봅시다. Petal_width를 먼저 그래프로 그려 보겠습니다.

```python
plt.plot(x_axis, iris['Petal_width']) 
plt.show()
```

그래프의 제목을 만들어 봅시다.

```python
plt.plot(x_axis, iris['Petal_width']) 
plt.title('Petal_width')
plt.show()
```

제목의 위치 및 크기를 변경해 봅시다.

```python
plt.plot(x_axis, iris['Petal_width']) 
plt.title('Petal_width', loc='right', pad=20, fontsize = 15)  # location
#loc : left/center/right, pad : 제목과 그래프 사이 간격, fontsize= title 크기
plt.show()
```

x축과 y축에 레이블을 표시해 봅니다.

```python
plt.plot(x_axis, iris['Petal_width']) 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.show()
```

x축과 y축 레이블과 그래프 사이 여백을 만들어 봅니다.

```python
plt.plot(x_axis, iris['Petal_width']) 
plt.title('Petal_width')
plt.xlabel('data', labelpad=10)
plt.ylabel('petal_width', labelpad=15)
plt.show()
```

범례를 표시해 봅시다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.legend(loc='lower right')  # loc 안 쓰면 알아서 디폴트로 위치 잡힘
# lower/upper, left/right, center 가능
# 튜플 형태도 가능. (0.0, 0.0): 왼쪽 하단,  (1.0,1.0): 오른쪽 상단
plt.show()
```

범례 크기를 키우고 꾸며 봅시다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.legend(loc='lower right', fontsize=14, frameon=True, shadow=True)  # frameon = False면 테두리가 없어진다
# frameon: 범례 테두리, shadow: 그림자
plt.show()
```

### 6.4. 축 조정하기

plt.xlim과 plt.ylim을 사용하여 축의 범위를 조절할 수 있습니다.

xmin: x축의 최소값, xmax: x축의 최대값, ymin: y축의 최소값, ymax: y축의 최대값

```python
plt.plot(x_axis, iris['Petal_width'], label='cm')
plt.title('Petal_width') 
plt.xlabel('data')
plt.ylabel('petal_width')
plt.xlim([0, 200]) # [xmin,xmax] 형태
plt.ylim([0, 3]) # [ymin, ymax] 형태
plt.show()
```

다른 방법으로, plt.axis([xim, xmax, ymin, ymax]) 형태로 줄 수도 있습니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.axis([0, 200, 0, 3])
plt.show()
```

역으로 그래프의 범위를 알고 싶으면 plt.xlim(), plt.ylim()을 호출하면 됩니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)
plt.show()
```

### 6.5. 선 그래프 꾸미기

간단한 기호들을 사용하여 선의 모양을 바꿀 수 있습니다.

'-' : Solid, '--' : Dashed, ':' : Dotted, '-.' : Dash-dot 

```python
plt.plot(x_axis, iris['Petal_width'], '--', label='cm')  # 디폴트가 - 하나다. 실선
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

linewidth를 사용해서 선의 굵기를 조절해 봅시다.

```python
plt.plot(x_axis, iris['Petal_width'], '--', linewidth=4, label='cm')
plt.title('Petal_width') 
plt.xlabel('data')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

데이터마다 점을 크게 만드는 마커를 만들 수 있습니다.

마커, 선 스타일, 색을 함께 표시하게끔 되어 있습니다.


Colors, Markers, Line Styles 순서대로 적어주면 됩니다.

ex) 빨간색 동그라미 마커와 점선 : 'ro--'

```python
plt.plot(x_axis, iris['Petal_width'], 'ro--', label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

argument를 직접 지정해서도 가능합니다.

```python
plt.plot(x_axis, iris['Petal_width'], color='red', marker='o', linestyle='--', label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

다양한 색깔, 선 종류, 마커 종류는 아래 표에 정리되어 있습니다.

image.png

특정 그래프의 영역을 채워 봅시다.

x축과 y축의 같은 slice를 입력하면 됩니다.

alpha는 투명도를 나타냅니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.fill_between(x_axis[10:40], iris['Petal_width'][10:40], alpha=0.4)  # x_axis 구간에 따른 영역 칠해서 채워줌
plt.show()
```

x축 방향의 그래프를 칠하고 싶으면 fill_betweenx()를 사용하면 됩니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.fill_betweenx(iris['Petal_width'][10:80], x_axis[10:80], alpha=0.4)
plt.show()
```

그리드를 설정해 볼수 있습니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.grid(True)
plt.show()
```

x축과 y축의 그리드만 설정할 수 있습니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.grid(True, axis='y')
plt.show()
```

grid도 선 그래프와 마찬가지로 스타일을 설정해 줄 수 있습니다.

```python
plt.plot(x_axis, iris['Petal_width'], label='cm') 
plt.title('Petal_width')
plt.xlabel('data')
plt.ylabel('petal_width')
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')
plt.show()
```

### 6.7 산점도 그리기

Petal_width와 Petal_length를 가지고 산점도를 그려 봅시다.

```python
plt.title('Petal_width and Petal_length')
plt.scatter(iris['Petal_length'], iris['Petal_width']) 
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

마커의 크기와 색상을 변경해 봅시다.

```python
plt.title('Petal_width and Petal_langth')
plt.scatter(iris['Petal_length'], iris['Petal_width'], s=20,c='green')  # size, color
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

Petal_width와 Petal_length, Sepal_width 를 가지고 3차원 산점도를 그려 봅시다.

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')  # 111 맨 앞 위치 그림 하나만 그리겠다
ax.scatter(iris['Petal_length'], iris['Petal_width'], iris['Sepal_width'], marker='o', s=15, c='darkgreen')
plt.title('Petal_width, Petal_langth, Sepal_width', pad=20)
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')
ax.set_zlabel('sepal_width')
plt.show()
```

여기에 라벨(Species_No) 까지 넣어서 산점도를 그려 봅시다.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig) 
labels = set(iris['Id'])
for i_ in labels:
  tt = iris[iris['Id']==i_]
  ax.scatter(tt['Petal_length'], tt['Petal_width'], tt['Sepal_width'], marker='o',s=35, label=i_)
plt.title('Petal_width, Petal_langth, Sepal_width', pad=20)
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')
ax.set_zlabel('sepal_width')
plt.show()
```
