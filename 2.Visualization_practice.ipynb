### 1. fmri 데이터

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
fmri['event'].unique()
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event')  # 신뢰구간 포함 (95% 신뢰구간 default 값이다)
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', err_style='bars')  # 각각의 신뢰구간까지 바 형태로 표시 가능하다
```

### 2. 펭귄 데이터

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

### barplot

```python
sns.barplot(data=penguins, x='island', y='body_mass_g', hue='sex')  # hue 는 쪼갠다
```

```python
sns.barplot(data=penguins, x='island', y='body_mass_g', hue='species')  # hue 는 쪼갠다
```

### countplot

```python
sns.countplot(x=penguins['island'])
```

```python
sns.countplot(x=penguins['species'])
```

### boxplot

```python
sns.boxplot(data=penguins, x='species', y='bill_length_mm')  # 각각의 종류별로 부리 길이에 대한 box_plot을 나타냄
```

```python
sns.boxplot(data=penguins, y='species', x='bill_length_mm')  # x와 y를 바꾸면 눕는다 (어차피 범주형/수치형 나뉘어져 있다)
```

```python
sns.boxplot(data=penguins, y='species', x='bill_length_mm', hue='sex')
```

### scatter plot

```python
sns.scatterplot(data=penguins, x='bill_depth_mm', y='body_mass_g')
```

```python
sns.scatterplot(data=penguins, x='bill_depth_mm', y='body_mass_g', hue='species')
```

```python
sns.scatterplot(data=penguins, x='bill_depth_mm', y='bill_length_mm', hue='species')
```

### violin plot

```python
sns.violinplot(data=penguins, x='bill_length_mm', y='island')
```

```python
sns.violinplot(data=penguins, x='bill_length_mm', y='island', hue='sex')
```

```python
sns.violinplot(data=penguins, x='bill_length_mm', y='island', hue='island')
```

### histplot

```python
sns.histplot(data=penguins, x='bill_length_mm')
```

```python
sns.histplot(data=penguins, x='bill_length_mm', binwidth=1)
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=5)
```

### kde (kernel density estimate) - hist와 비슷하나 연속함수로 근사

```python
sns.kdeplot(data=penguins, x='bill_length_mm')
```

```python
sns.kdeplot(data=penguins, x='bill_length_mm', bw_adjust=0.2)  # 가우시안 값을 더 민감하게 뽑을 수 있다 
```

### 3. glue data

```python
glue = sns.load_dataset('glue')
```

```python
glue.head()
```

### pivot

```python
df = glue.pivot('Model', 'Task', 'Score')  # 행, 열, 값
```

```python
df
```

### heatmap

```python
sns.heatmap(data=df, annot=True, fmt='.1f', linewidth=0.5)
```
